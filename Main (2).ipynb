## XML to CSV file coverter

I used minidom pakage to take out the data from the given tags and store it in xmltag format

```python
from xml.dom import minidom

xmldoc = minidom.parse('DLTINS_20210117_01of01.xml')
id_lst = xmldoc.getElementsByTagName('Id')
FullNm_lst = xmldoc.getElementsByTagName('FullNm')
ClssfctnTp_lst = xmldoc.getElementsByTagName('ClssfctnTp')
CmmdtyDerivInd_lst= xmldoc.getElementsByTagName('CmmdtyDerivInd')
NtnlCcy_lst= xmldoc.getElementsByTagName('NtnlCcy')
Issr_lst = xmldoc.getElementsByTagName('Issr')
```

```python
len(id_lst),len(FullNm_lst) , len(ClssfctnTp_lst) , len(CmmdtyDerivInd_lst) , len(NtnlCcy_lst) , len(Issr_lst)
```

created a funtion to covert xmltag format storage to simple list for so that we can interpret that data

```python
def cvt(obj):
    store = []
    for i in range(0,len(obj)):
        store.append(obj[i].firstChild.nodeValue)
    return store
```

```python
id_lst= cvt(id_lst)
FullNm_lst = cvt(FullNm_lst)
ClssfctnTp_lst = cvt(ClssfctnTp_lst)
CmmdtyDerivInd_lst = cvt(CmmdtyDerivInd_lst)
NtnlCcy_lst = cvt(NtnlCcy_lst)
Issr_lst = cvt(Issr_lst)
```

Imported it to the csv with the required name 

```python
import csv

col = ['FinInstrmGnlAttrbts.Id',
'FinInstrmGnlAttrbts.FullNm',
'FinInstrmGnlAttrbts.ClssfctnTp',
'FinInstrmGnlAttrbts.CmmdtyDerivInd',
'FinInstrmGnlAttrbts.NtnlCcy',"Issr"]

output_data = open('output.csv' , 'w' , newline='' , encoding='utf-8')
csvwriter = csv.DictWriter(output_data , fieldnames= col)

```

I got many extra id's due to present of ID tag in every tag

```python
csvwriter.writeheader()
for i in range(0 ,14138):
    csvwriter.writerow({'FinInstrmGnlAttrbts.Id' : id_lst[i],
'FinInstrmGnlAttrbts.FullNm':FullNm_lst[i],
'FinInstrmGnlAttrbts.ClssfctnTp':ClssfctnTp_lst[i] ,
'FinInstrmGnlAttrbts.CmmdtyDerivInd': CmmdtyDerivInd_lst[i] ,
'FinInstrmGnlAttrbts.NtnlCcy': NtnlCcy_lst[i],
 'Issr' : Issr_lst[i]})
```
